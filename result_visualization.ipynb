# Result Visualization

This file plots experiment run results for analysis.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_ttt = pd.read_csv("./run_results/ttt.csv", converters={'session_timestamp': str})
df_ttt
```

```python
df_con4 = pd.read_csv("./run_results/con4.csv", converters={'session_timestamp': str})
df_con4
```

## Experiment 1

```python
df_exp1 = df_ttt[df_ttt['session_id'].str.contains('exp1')]
df_exp1_minimax = df_exp1[df_exp1['player2'] == "StrategyMiniMax"]
df_exp1_tabq = df_exp1[df_exp1['player2'] == "StrategyTabQLearning"]

performance_summary = []

# Get performance of minimax agent.
win_rate = (sum(df_exp1_minimax['outcome'] == 2)/len(df_exp1_minimax))*100
loss_rate = (sum(df_exp1_minimax['outcome'] == 1)/len(df_exp1_minimax))*100
draw_rate = (sum(df_exp1_minimax['outcome'] == 0)/len(df_exp1_minimax))*100
avg_move_ms = df_exp1_minimax['avg_milliseconds_per_move_player2'].mean()
avg_game_moves = df_exp1_minimax['num_moves'].mean()
ps_exp1_minimax = {}
ps_exp1_minimax['experiment'] = 1
ps_exp1_minimax['game'] = 'Tic Tac Toe'
ps_exp1_minimax['player'] = 'Minimax (Alpha Beta)'
ps_exp1_minimax['opponent'] = 'default'
ps_exp1_minimax['win_rate'] = win_rate
ps_exp1_minimax['loss_rate'] = loss_rate
ps_exp1_minimax['draw_rate'] = draw_rate
ps_exp1_minimax['avg_move_ms'] = avg_move_ms
ps_exp1_minimax['avg_game_moves'] = avg_game_moves
performance_summary.append(ps_exp1_minimax)

# Get performance of q learning agent.
win_rate = (sum(df_exp1_tabq['outcome'] == 2)/len(df_exp1_tabq))*100
loss_rate = (sum(df_exp1_tabq['outcome'] == 1)/len(df_exp1_tabq))*100
draw_rate = (sum(df_exp1_tabq['outcome'] == 0)/len(df_exp1_tabq))*100
avg_move_ms = df_exp1_tabq['avg_milliseconds_per_move_player2'].mean()
avg_game_moves = df_exp1_tabq['num_moves'].mean()
ps_exp1_tabq = {}
ps_exp1_tabq['experiment'] = 1
ps_exp1_tabq['game'] = 'Tic Tac Toe'
ps_exp1_tabq['player'] = 'Tabular Q Learning'
ps_exp1_tabq['opponent'] = 'default'
ps_exp1_tabq['win_rate'] = win_rate
ps_exp1_tabq['loss_rate'] = loss_rate
ps_exp1_tabq['draw_rate'] = draw_rate
ps_exp1_tabq['avg_move_ms'] = avg_move_ms
ps_exp1_tabq['avg_game_moves'] = avg_game_moves
performance_summary.append(ps_exp1_tabq)

df_exp1_ps = pd.DataFrame(performance_summary)
df_exp1_ps
```
